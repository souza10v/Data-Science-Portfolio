```python
#Importando bibliotecas

#Manipulacao de dados
import numpy as np
import pandas as pd

#Visualizacao de dados
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

# machine learning
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC, LinearSVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import Perceptron
from sklearn.linear_model import SGDClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder
```

```python
spaceship_train_df =  pd.read_csv('train.csv')
spaceship_test_df = pd.read_csv('test.csv')
```

```python
spaceship_train_df.isna()
```

```python
spaceship_train_df = spaceship_train_df.fillna(-1)
spaceship_test_df = spaceship_test_df.fillna(-1)
```

```python
def age_classification(passanger):
  age = passanger.Age
  if age < 18:
    return 'Child'
  elif age >= 18 and age <=60:
    return 'Adult'
  else:
    return 'Old'

def get_cabin(passanger, aux):
  cabin_num = str(passanger)
  if cabin_num != 'nan' and cabin_num != '-1' :
    cabin_num = cabin_num.split('/')
    return cabin_num[aux]
  else:
    return(cabin_num)
```

```python
spaceship_train_df['Person'] = spaceship_train_df.apply(age_classification, axis=1)
spaceship_train_df['Deck'] = spaceship_train_df.Cabin.apply(lambda x: get_cabin(x, 0))
spaceship_train_df['Num'] = spaceship_train_df.Cabin.apply(lambda x: get_cabin(x, 1))
spaceship_train_df['Side'] = spaceship_train_df.Cabin.apply(lambda x: get_cabin(x, 2))

spaceship_test_df['Person'] = spaceship_test_df.apply(age_classification, axis=1)
spaceship_test_df['Deck'] = spaceship_test_df.Cabin.apply(lambda x: get_cabin(x, 0))
spaceship_test_df['Num'] = spaceship_test_df.Cabin.apply(lambda x: get_cabin(x, 1))
spaceship_test_df['Side'] = spaceship_test_df.Cabin.apply(lambda x: get_cabin(x, 2))
```

```python
spaceship_train_df['Transported'] = spaceship_train_df['Transported'].map({True: 1, False: 0})
spaceship_train_df['CryoSleep'] = spaceship_train_df['CryoSleep'].map({True: 1, False: 0})
spaceship_train_df['VIP'] = spaceship_train_df['VIP'].map({True: 1, False: 0})
spaceship_train_df['HomePlanet'] = spaceship_train_df['HomePlanet'].map({'Europa': 0, 'Earth': 1, 'Mars': 2})
spaceship_train_df['Person'] = spaceship_train_df['Person'].map({'Child': 0, 'Adult': 1, 'Old': 2})
spaceship_train_df['Destination'] = spaceship_train_df['Destination'].map({'TRAPPIST-1e': 0, 'PSO J318.5-22': 1, '55 Cancri e': 2})
spaceship_train_df['Deck'] = spaceship_train_df['Deck'].map({'A': 0, 'B': 1, 'C': 2, 'D': 2, 'E': 2, 'F': 2, 'G': 2, 'T': 2})
spaceship_train_df['Side'] = spaceship_train_df['Side'].map({'P': 0, 'S': 1})

spaceship_test_df['CryoSleep'] = spaceship_test_df['CryoSleep'].map({True: 1, False: 0})
spaceship_test_df['VIP'] = spaceship_test_df['VIP'].map({True: 1, False: 0})
spaceship_test_df['HomePlanet'] = spaceship_test_df['HomePlanet'].map({'Europa': 0, 'Earth': 1, 'Mars': 2})
spaceship_test_df['Person'] = spaceship_test_df['Person'].map({'Child': 0, 'Adult': 1, 'Old': 2})
spaceship_test_df['Destination'] = spaceship_test_df['Destination'].map({'TRAPPIST-1e': 0, 'PSO J318.5-22': 1, '55 Cancri e': 2})
spaceship_test_df['Deck'] = spaceship_test_df['Deck'].map({'A': 0, 'B': 1, 'C': 2, 'D': 2, 'E': 2, 'F': 2, 'G': 2, 'T': 2})
spaceship_test_df['Side'] = spaceship_test_df['Side'].map({'P': 0, 'S': 1})
```

```python
header_x = ['HomePlanet', 'CryoSleep', 'Destination', 'Age',
       'VIP', 'RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck', 
       'Person', 'Deck', 'Num', 'Side']

X_train = spaceship_train_df[header_x]
Y_train = spaceship_train_df.Transported

X_train = X_train.fillna(-1)

X_test = spaceship_test_df[header_x]
X_test = X_test.fillna(-1)
```

```python
X_train.shape, Y_train.shape, X_test.shape
```

```python
 def aux_transfor_(x):
  array = []
  for i in x:
    if i == 0:
      array.append(False)
    else: 
      array.append(True)
  return np.asarray(array)   
    
```

Logistic Regression

```python
logreg = LogisticRegression()
logreg.fit(X_train, Y_train)
Y_pred = logreg.predict(X_test)
acc_log = round(logreg.score(X_train, Y_train) * 100, 2)
Y_pred = aux_transfor_(Y_pred)
acc_log
```

```python
sub = pd.Series(Y_pred ,index=spaceship_test_df['PassengerId'], name='Transported')
sub.to_csv('LogisticRegression.csv', header=True)
!head -n10 LogisticRegression.csv
```
