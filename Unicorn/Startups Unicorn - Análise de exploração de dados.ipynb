```python
unicorn_df['Date Joined'] = pd.to_datetime(unicorn_df['Date Joined'])
```

```python
unicorn_df['Month'] = pd.DatetimeIndex(unicorn_df['Date Joined']).month
unicorn_df['Year'] = pd.DatetimeIndex(unicorn_df['Date Joined']).year
```

```python
base_agrupada = unicorn_df.groupby(by = ['Country', 'Year', 'Month', 'Company']).count()['Date Joined'].reset_index()
```

```python
unicorn_df['Valuation ($B)'] = unicorn_df['Valuation ($B)'].str.replace("$", "").astype(float)
```

```python
unicorn_df
```

```python
analise_paises = unicorn_df.groupby(by = ['Country']).sum()['Valuation ($B)'].reset_index()
```

```python
analise_valuation = analise_paises.sort_values('Valuation ($B)', ascending = False)
```

```python
plt.figure(figsize = (15,6))
plt.plot(analise_valuation['Country'], analise_valuation['Valuation ($B)'])
plt.xticks(rotation=45, ha='right');
```
