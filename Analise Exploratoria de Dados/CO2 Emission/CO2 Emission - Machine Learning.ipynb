```python
#Importando bibliotecas

#Manipulacao de dados
import numpy as np
import pandas as pd

#Visualizacao de dados
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import missingno

#Sklearn
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC, LinearSVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import Perceptron
from sklearn.linear_model import SGDClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split

import warnings
warnings.filterwarnings('ignore')
```

```python
co2_consumption = pd.read_csv('FuelConsumptionCo2.csv')
```

<h1>Machine Learning Pre Processamento</h1>

```python
labelencoder1 = LabelEncoder()
co2_consumption['MAKE'] = labelencoder1.fit_transform(co2_consumption['MAKE'])

labelencoder2 = LabelEncoder()
co2_consumption['MODEL'] = labelencoder2.fit_transform(co2_consumption['MODEL'])

labelencoder3 = LabelEncoder()
co2_consumption['VEHICLECLASS'] = labelencoder3.fit_transform(co2_consumption['VEHICLECLASS'])

labelencoder4 = LabelEncoder()
co2_consumption['TRANSMISSION'] = labelencoder4.fit_transform(co2_consumption['TRANSMISSION'])

labelencoder5 = LabelEncoder()
co2_consumption['FUELTYPE'] = labelencoder5.fit_transform(co2_consumption['FUELTYPE'])
```

```python
co2_consumption.head()
```

```python
x = co2_consumption.drop('CO2EMISSIONS', axis=1)
y = co2_consumption['CO2EMISSIONS']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3)
```
