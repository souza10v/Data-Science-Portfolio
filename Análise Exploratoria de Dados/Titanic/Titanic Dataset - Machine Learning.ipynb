```python
#Importando bibliotecas

#Manipulacao de dados
import numpy as np
import pandas as pd

#Visualizacao de dados
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

# machine learning
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC, LinearSVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import Perceptron
from sklearn.linear_model import SGDClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder
```

```python
titanic_train_df = pd.read_csv('train.csv')
titanic_test_df = pd.read_csv('test.csv')
```

```python
titanic_train_df['Age_bin'] = pd.cut(titanic_train_df['Age'], bins=[0,12,20,40,120], labels=['Children','Teenage','Adult','Elder'])
titanic_train_df['Fare_bin'] = pd.cut(titanic_test_df['Fare'], bins=[0,7.91,14.45,31,120], labels=['Low_fare','median_fare', 'Average_fare','high_fare'])
                                                                                                   
titanic_test_df['Age_bin'] = pd.cut(titanic_test_df['Age'], bins=[0,12,20,40,120], labels=['Children','Teenage','Adult','Elder'])
titanic_test_df['Fare_bin'] = pd.cut(titanic_test_df['Fare'], bins=[0,7.91,14.45,31,120], labels=['Low_fare','median_fare', 'Average_fare','high_fare'])                                                                                                   
```

```python
header_x = [ 'Pclass', 'Sex', 'Age_bin', 'SibSp',
       'Parch', 'Fare_bin', 'Embarked']

labelencoder = LabelEncoder()
titanic_train_df.Sex = labelencoder.fit_transform(titanic_train_df.Sex)
titanic_train_df.Embarked = labelencoder.fit_transform(titanic_train_df.Embarked)
titanic_train_df.Age_bin = labelencoder.fit_transform(titanic_train_df.Age_bin)
titanic_train_df.Fare_bin = labelencoder.fit_transform(titanic_train_df.Fare_bin)

titanic_test_df.Sex = labelencoder.fit_transform(titanic_test_df.Sex)
titanic_test_df.Embarked = labelencoder.fit_transform(titanic_test_df.Embarked)
titanic_test_df.Age_bin = labelencoder.fit_transform(titanic_test_df.Age_bin)
titanic_test_df.Fare_bin = labelencoder.fit_transform(titanic_test_df.Fare_bin)

X_train = titanic_train_df[header_x]
Y_train = titanic_train_df.Survived

X_train = X_train.fillna(-1)

X_test = titanic_test_df[header_x]
X_test = X_test.fillna(-1)
```

```python
X_train.shape, Y_train.shape, X_test.shape
```

Logistic Regression

```python
logreg = LogisticRegression()
logreg.fit(X_train, Y_train)
Y_pred = logreg.predict(X_test)
acc_log = round(logreg.score(X_train, Y_train) * 100, 2)
acc_log
```

```python
sub = pd.Series(Y_pred ,index=titanic_test_df['PassengerId'], name='Survived')
sub.to_csv('LogisticRegression.csv', header=True)
!head -n10 LogisticRegression.csv
```
